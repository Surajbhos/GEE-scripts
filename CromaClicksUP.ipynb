```python
import os.path

def generate_sections_of_url(url):
    path = urllib.parse.urlparse(url).path
    sections = []; temp = "";
    while path != '/':
        temp = os.path.split(path)
        path = temp[0]
        sections.append(temp[1])
    return sections
```

```python
def extract_path_section(url):
    path = urlparse(url).path
    sections = path.split('/')
    sections = sections[1:]
    que = urlparse(url).query
    return sections
```

```python
def extract_query(url):
    que = urlparse(url).query
    return que
```

### Function to create New Timestamp Column

```python
# Function to create New Timestamp Column

def create_timest(datafm):
      # convert the datetime column to a pandas datetime object
      datafm['Time Stamp (prop3)'] = pd.to_datetime(datafm['Time Stamp (prop3)'])
      print(datafm.dtypes)
      # convert the datetime column to an integer
      datafm['New Timestamp'] = datafm['Time Stamp (prop3)'].astype(int)
      # divide the resulting integer by the number of nanoseconds in a second
      datafm['New Timestamp'] = datafm['New Timestamp'].div(10**9)
```

### Function to return count of unique and duplicate records in dataframe. It takes dataframe as input

```python

def uniq(dat):
  a = len(dat)
  d = len(duplicate)
  b = len(dat.drop_duplicates(subset=['New Timestamp']))
  #count number of duplicate rows
  c = a-b
  e = a-d
  print("Duplicate Rows:", d)
  print("Unique rows:", e)
  print("length of dataframe:", a)
  print("length of records with Distinct timestamp:", b)
  print("Seperated duplicates:", c)
```
